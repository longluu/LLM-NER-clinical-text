```python
from src.data.data_loader import *
import pandas as pd

dataset_loader = DatasetLoader(dataset_name='../data/public/MedMentions/preprocessed-data/', path_umls_semtype='../data/public/MedMentions/SemGroups_2018.txt', model_name='UFNLP/gatortron-base')
data_medmentions, classmap, umls_label_code, tokenizer = dataset_loader.load_dataset()
```

```python
data_medmentions
```

```python
ind_example = 0
df = pd.DataFrame(columns=['tokens', 'ner_tags', 'token_labels', 'umls_label_code'])
df['tokens'] = data_medmentions['train'][ind_example]['tokens']
df['token_labels'] = data_medmentions['train'][ind_example]['token_labels']
df['ner_tags'] = data_medmentions['train'][ind_example]['ner_tags']
df['umls_label_code'] = [umls_label_code[code] for code in data_medmentions['train'][ind_example]['ner_tags']]
df.T
```

```python
df = pd.DataFrame(columns=['input_ids', 'input', 'labels'])
df['input_ids'] = data_medmentions['train'][ind_example]['input_ids']
df['input'] = [tokenizer.decode([t]) for t in tokenizer.encode(' '.join(data_medmentions['train'][ind_example]['tokens']))]
df['labels'] = data_medmentions['train'][ind_example]['labels']
df.T
```
